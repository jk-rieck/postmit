```python
import xarray as xr
import postmit as pm
```

```python
path_to_data = "/path/to/data/"
```

### Load some data
Here we load data that has already been converted to `netCDF` format with [`convert.convert2nc`](https://github.com/jk-rieck/postmit/blob/main/examples/example_convert.ipynb).

```python
ds = xr.open_dataset(path_to_data + "data_1m_0222.nc")
```

### Apply some variables
Here we will add latitude and longitude to a simulation with a cartesian grid with [__`add_lat_lon`__](https://github.com/jk-rieck/postmit/blob/main/postmit/adds.py#L8). This is arbitrary and might not make sense in all cases. However, if for example the atmospheric forcing used is based on real-world data (i.e. can be referenced to latitudes and longitudes) it can be helpful to know the approximate 'lat-lon' coordinates that a point in the simulation corresponds to.

```python
ds = pm.adds.add_lat_lon(ds, latmin=-75, latmax=-35, lonmin=30, lonmax=30)
```

This call adds latitude and longitude variables for all four grid locations ((`XC`, `YC`), (`XC`, `YG`), (`XG`, `YC`), (`XG`, `YG`)). If the inputs `latmin` and `latmax` for latitude (`lonmin` and `lonmax` for longitude) are identical, it is assumed that one single latitude (longitude) is adequate for the entire domain.  
See for example below, where `latC` varies with `YC`, whereas `lonC` is the same value (`30`) everywhere.

```python
ds.latC.plot();
```

```python
ds.lonC.plot();
```
